```python
import us
import pandas as pd
import numpy as np
import requests
import json
```

```python
states = us.states.STATES
state_abbr = []

for state in states:
    state_abbr.append(state.abbr)

state_abbr = [name.lower() for name in state_abbr]

covid_deaths_2021_df = pd.DataFrame({'State': states, 'State Abbreviation':state_abbr})
covid_deaths_2021_df ["Deaths 2021-12-31"] = ""
covid_deaths_2021_df ["Deaths 2021-01-01"] = ""
covid_deaths_2021_df ["Total Deaths 2021"] = ""
covid_deaths_2021_df
```

```python
url = "https://api.covidtracking.com/v1/states/"

for index, row in covid_deaths_2021_df.iterrows():
    query_url = f"{url}{row['State Abbreviation']}/daily.json"
    response = requests.get(query_url).json()
    covid_deaths_2021_df.loc[index, "Deaths 2021-12-31"] = response[0]["death"]
    covid_deaths_2021_df.loc[index, "Deaths 2021-01-01"] = response[0]["death"]

#find out how to get death data by date (if/try statement?)

covid_deaths_2021_df
```
